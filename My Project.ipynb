# Data Cleaning

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('C:/Users/MANAR/Downloads/Salaries.csv')
print(df)
```

```python
# Drop rows with any missing values
df_dropped = df.dropna()
print(df_dropped)
```

```python
# Drop columns with any missing values
df_dropped = df.dropna(axis=1)
print(df_dropped)
```

```python
# Impute missing values with the mean of each column
df_imputed_mean = df.fillna(df.mean())
print(df_imputed_mean)
```

Imputing missing values with mean:

The dataset contains numerical values such as BasePay, OvertimePay, OtherPay, Benefits, TotalPay, and TotalPayBenefits.

Imputing missing values with the mean is a common method for numerical data when the missing values are assumed to be missing at random.

By replacing missing values with the mean, we can maintain the overall statistical properties of the data.

This method is relatively simple and quick to implement, and it helps retain the sample size and consistency in subsequent calculations or analyses.

Additionally, imputing with the mean can be a reasonable approach if the missing data is relatively small compared to the total dataset and does not significantly affect the analysis.


# Basic Data Visualization

```python
#visualize the distribution of salaries
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(df['BasePay'], bins=10)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries')
plt.show()

# Create a bar chart to represent the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts().head(10)
plt.bar(department_counts.index, department_counts.values)
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Proportion of Employees in Different Departments')
plt.xticks(rotation=90)
plt.show()
```

```python
#the proportion of employees in different departments 
department_counts = df['JobTitle'].value_counts()
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments')
plt.axis('equal')
plt.show()
```

# Grouped Analysis

```python
average_salary_by_department = df.groupby('JobTitle')['BasePay'].mean()
print(average_salary_by_department)
```

```python
#compare the average salaries across different groups
correlation = df['BasePay'].corr(df['OvertimePay'])
print("Correlation between BasePay and OvertimePay:", correlation)

# Scatter plot
plt.scatter(df['BasePay'], df['OvertimePay'])
plt.xlabel('BasePay')
plt.ylabel('OvertimePay')
plt.title('Scatter Plot: BasePay vs. OvertimePay')
plt.show()
```

# Simple Correlation Analysis

```python
# Calculate the correlation coefficient between salary and BasePay
correlation = df['TotalPayBenefits'].corr(df['BasePay'])
print("Correlation between Salary and BasePay:", correlation)

# Plot a scatter plot to visualize the relationship between salary and BasePay
plt.scatter(df['BasePay'], df['TotalPayBenefits'])
plt.xlabel('BasePay')
plt.ylabel('Salary')
plt.title('Relationship between Salary and BasePay')
plt.show()
```
